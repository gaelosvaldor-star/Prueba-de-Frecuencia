Prueba de los promedios

A. El Promedio (x_barra)
Es simplemente la suma de todos tus números dividida entre cuántos hay (N).
x_barra = sum(xi) / N

B. El Estadístico de Prueba (Z0)
Fórmula estándar corregida:
Z0 = (x_barra - 0.5) * sqrt(12N)

C. El Criterio de Decisión
Comparamos el valor absoluto de nuestro Z0 contra un valor de la tabla normal (Z_alpha/2).

  - Si |Z0| < Z_alpha/2: ¡Pasa la prueba! (No se rechaza H0).
  - Si |Z0| >= Z_alpha/2: No son uniformes (Se rechaza H0)

```python
alpha = float(input("Ingrese alpha: "))
def prueba_prom():
    n = len(lista_1)
    x_prom = st.mean(lista_1)

    z0 = (x_prom - 0.5) * np.sqrt(12 * n)

    print(f"--- Análisis ---")
    print(f"Z0 calculado: {abs(z0):.4f}")

    if np.abs(z0) < alpha:
        print("Resultado: La hipótesis nula NO se rechaza (Distribución Uniforme)")
    else:
        print("Resultado: La hipótesis nula SE RECHAZA (No es Uniforme)")

prueba_prom()
```
